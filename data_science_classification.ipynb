```python
"""Skewness is a measure of asymmetry of a distribution.
A skewness value of 0 in the output denotes a symmetrical distribution of values in row 1.
A negative skewness value in the output indicates an asymmetry in the distribution 
and the tail is larger towards the left hand side of the distribution. A positive skewness 
value in the output also indicates an asymmetry in the distribution and the tail 
is larger towards the right hand side of the distribution.
· If the skewness is between -0.5 and 0.5, the data are fairly symmetrical
· If the skewness is between -1 and — 0.5 or between 0.5 and 1, the data are moderately skewed
· If the skewness is less than -1 or greater than 1, the data are highly skewed
"""
#skew_value = satisfied.skew(axis=1)
skew_value = satisfied.select_dtypes(include='number').skew(axis=1)
```

```python
skew_value
```

```python
"""Kurtosis describes the peakedness of the distribution.
If the distribution is tall and thin it is called a leptokurtic distribution(Kurtosis > 3). 
Values in a leptokurtic distribution are near the mean or at the extremes.
A flat distribution where the values are moderately spread out (i.e., unlike leptokurtic) 
is called platykurtic(Kurtosis <3) distribution. A distribution whose shape is in between 
a leptokurtic distribution and a platykurtic distribution is called a mesokurtic(Kurtosis=3) distribution. 
A mesokurtic distribution looks more close to a normal distribution.
Skewness essentially measures the symmetry of the distribution, 
while kurtosis determines the heaviness of the distribution tails.
· It is the sharpness of the peak of a frequency-distribution curve .
It is actually the measure of outliers present in the distribution.
· High kurtosis in a data set is an indicator that data has heavy outliers.
· Low kurtosis in a data set is an indicator that data has lack of outliers.
"""
#kurtosis_value = satisfied.kurt(axis=1)
kurtosis_value = satisfied.select_dtypes(include='number').kurt(axis=1)

kurtosis_value
```

# Machine Learning for customer classification [Satisfied, Unsatisfied]

```python
wifi["Satisfaction"].replace(to_replace ="Satisfied",
                 value = 0, inplace=True)
wifi["Satisfaction"].replace(to_replace ="Unsatisfied",
                 value = 1, inplace=True)
wifi["Satisfaction"].replace(to_replace ="Somewhat Satisfied",
                 value = 2, inplace=True)
wifi
```

# Drop "somewhat satisfied" cases and related features from dataset

```python
labels = wifi.iloc[1:,:]
trainset = wifi.iloc[1:,1:]
trainset
```

# Drop "Observation" field

```python
trainset = trainset.drop('Observation', axis=1)

trainset
```

```python
from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler
scaler = RobustScaler()

scaled_rows = scaler.fit_transform(trainset.T).T
scaled_df = pd.DataFrame(data = scaled_rows, columns = trainset.columns)
scaled_df
```

```python
index = 3
print(scaled_df.iloc[index][:])
scaled_df.iloc[index][:].plot(figsize=(15,8))
```

```python
from sklearn.model_selection import train_test_split
X = scaled_df
y = labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, random_state=0)
#clf = svm.SVC(gamma = 0.1, kernel="rbf", C=10)
#clf.fit(X_train,y_train)
clf.fit(X_train, y_train.values.ravel())

y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report

mean_squared_error(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred, labels=[0, 1]))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix,plot_confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

plt.figure(figsize=(10, 10))
ConfusionMatrixDisplay.from_estimator(
    clf, 
    X_test, 
    y_test, 
    cmap=plt.cm.Blues
)
plt.title("Confusion Matrix")
plt.savefig("confusion-matrix.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
CM = confusion_matrix(y_test, y_pred)

TN = CM[0][0]
FN = CM[1][0]
TP = CM[1][1]
FP = CM[0][1]

# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/(TP+FN)
# Specificity or true negative rate
TNR = TN/(TN+FP) 
# Precision or positive predictive value
PPV = TP/(TP+FP)
# Negative predictive value
NPV = TN/(TN+FN)
# Fall out or false positive rate
FPR = FP/(FP+TN)
# False negative rate
FNR = FN/(TP+FN)
# False discovery rate
FDR = FP/(TP+FP)

# Overall accuracy
ACC = (TP+TN)/(TP+FP+FN+TN)
```

```python
print("ACC:", ACC, 
      "TPR:", TPR, 
      "TNR:", TNR, 
      "PPV:", PPV, 
      "NPV:", NPV, 
      "FPR:", FPR, 
      "FNR:", FNR, 
      "FDR:", FDR)
```

# Extra feature extraction from fields for achieving higher accuracy in machine learning 

```python
trainset = trainset.reset_index(drop=True)
trainset = trainset.set_index(trainset.index)
trainset
```

#  Shannon Entropies as feature 

```python
from scipy.stats import entropy
entropies_for_each_row = []
for i in range(len(trainset)):
    E = entropy(trainset.iloc[i][2:].astype(float), base=2)
    entropies_for_each_row.append(E)
entropies_for_each_row
```

```python
entropies = pd.DataFrame(entropies_for_each_row)
entropies.columns = ["Entropy"]
entropies = entropies.reset_index(drop = True)
entropies = entropies.set_index(entropies.index)
entropies
```

```python
entropies["Entropy"].plot(figsize=(16,6))
```

# Power Spectral Density as feature

```python
import numpy as np
from scipy import signal

specs_for_each_row = []

for i in range(len(trainset)):
    sig = trainset.iloc[i][2:]  
    freqs, Pxx = signal.periodogram(sig.astype(float))  

    specs_for_each_row.append(Pxx.mean())

norm = np.linalg.norm(specs_for_each_row)
specs_for_each_row = np.array(specs_for_each_row) / norm

specs_for_each_row
```

```python
specs = pd.DataFrame(specs_for_each_row)
specs.columns = ["Normalized Specs"]
specs = specs.reset_index(drop = True)
specs = specs.set_index(specs.index)
specs
```

```python
specs["Normalized Specs"].plot(figsize=(16,6))
```

```python
sig = trainset.iloc[0][2:] 
freqs, Pxx = signal.periodogram(sig.astype(float)) 

plt.figure(figsize=(8, 6))
plt.semilogy(freqs, Pxx)  
plt.title('Power Spectral Density (Periodogram) - First Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.tight_layout()
plt.show()
```
